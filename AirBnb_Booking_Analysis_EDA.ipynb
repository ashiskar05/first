```python
ax=sns.countplot(x=df["room_type"])
ax.set_title("Room_Type")
plt.show();
```

**Entire home or apt is most favored room type and shared room is least favored.â€‹**

```python
#  find the maximum(minimum nights) spending by customer across different room type
total_room=df.groupby('room_type')['minimum_nights'].count().reset_index()
room_types=total_room.sort_values('minimum_nights',ascending=True)
room_types.head()
```

```python
# create dataset
labels=list(room_types['room_type'])
sizes=list(room_types['minimum_nights'])
# create color parameter
colors=['green','orange','red']
# create explode
explode = (0.07, 0.07, 0.07)
```

```python
#creating pie chart
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.title('Percentage of nights spend per room types', fontsize=20)
plt.axis("equal")
plt.show()
```

```python
#Get the host based on number of reviews
host_based_on_review_df = df.groupby(['host_name'],as_index=False)['number_of_reviews'].sum().sort_values(['number_of_reviews'],ascending = False)
host_based_on_review_df
```

```python
#Get the host based on availability in a year
host_based_on_availability_df = df.groupby(['host_name'],as_index=False)['availability_365'].mean().sort_values(['availability_365'],ascending = True)
host_based_on_availability_df
```

```python
#Merge two dataframe based on number of reviews and availability in a year
popular_host_df = pd.merge(host_based_on_review_df,host_based_on_availability_df,on = ['host_name']).sort_values(['availability_365','number_of_reviews'],ascending = [True,False]).rename(columns = {'number_of_reviews':'Number of reviews','availability_365':'Availability in a year'})

popular_host_df
```

```python
#Most popular top 5 host in new york
popular_host_df.head(5)
```

```python
#Create scatter plot based on number of reviews and availability in a year
popular_host_df.plot.scatter(x='Number of reviews', y='Availability in a year', xlim=(100,1000),ylim=(10,380));
```

```python
plt.figure(figsize=(10,10))
sns.set_style('white')

ax = sns.boxplot(data=df, x='neighbourhood_group',y='availability_365',palette='plasma')

# Naming the Chart
ax.set_title('Relation between Neighbourhood group & Availability of rooms').set_fontsize('15')

# Naming X & Y axis
ax.set_ylabel('Availability 365 Days').set_fontsize('15')
ax.set_xlabel('Neighbourhood Group').set_fontsize('15')

#Adjusting Bar labels
ax.set_xticklabels(ax.get_xticklabels(), size = '15')
```

```python
#function to catogory the types of rooms.
def price_catagory(price):
  if price<50:
    return 'cheap'
  elif price>=50 and price<=200:
    return 'affordable'
  else:
    return 'Expensive'
```

```python
plt.figure(figsize=(10,5))
ax= sns.countplot(x=df['price'].apply(price_catagory))
ax.set_title('Count Plot')
ax.set_xlabel('Catogories of rooms')
ax.set_ylabel("Count")
plt.show()
```

```python
(df['price'].apply(price_catagory).value_counts(normalize=True)).round(2)
```

#### Bi-Variate Analysis

```python
#Average price of room in different neighbourhood group
locationwiseprice_data = df.groupby(['neighbourhood_group','room_type'])['price'].mean().unstack()
locationwiseprice_data.plot(kind='bar',figsize=(10,6))
plt.title('Average Price of different types of room in every neighbourhood group')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (8,5))
ax= sns.barplot(y= 'price', x= 'room_type',data=df)
ax.set_title('Price vs Room Type')
ax.set_xlabel('Room Type')
ax.set_ylabel("Price")
plt.show()
```

**Average price of entire home/apt is highest in Manhattan.**
